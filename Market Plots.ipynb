```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sensex = pd.read_csv("Sensex_52week.csv")
```

```python
ax=sns.lineplot(x="Date", y="Close", data=sensex, color='green')
ax.set_xticks([0,50,100,150,200,243])
plt.title("52-Week Sensex")
plt.ylabel("Closing Value")
sns.set_style("whitegrid")
```
